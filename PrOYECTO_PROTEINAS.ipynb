# Duplicación
El  proceso  de  duplicación  de  ADN  es  el  mecanismo  que  permite  al DNA  replicarse  (es  decir,  sintetizar  una  copia  idéntica).
La  duplicación  empieza  en  puntos  determinados:  los  orígenes  de replicación.   Las   proteínas   iniciadoras   reconocen   secuencias   denucleótidos  específicas  en  esos  puntos  y  facilitan  la  fijación  de  otras proteínas  que  permitirán  la  separación  de  las  dos  hebras  de  DNA formándose una horquilla de replicación. Para representar el proceso se generó un código que toma la cadena original de DNA perteneciente a una patología que con un ciclo for se asignaran las basaes nitrogenadas correspondientes a la cadena original creando la cadena hija.



```python
f= open(archivo, "r") #abre el archivo seleccionado arriba en modo lectura
tex=f.read()  #se lee el archivo
ADN=tex  #define la variable
#ADN= input(n)
nuevo=" "  #creaciòn de "nuevo" (cadena vacía) a la cual llegarán los nucleotidos generados por el ciclo for de DNA
for i in ADN: #para cada i en DNA
    if i=='A':  #si i es igual a A entonces 
        nuevo +="\033[1;31m"+'T' # se generará en "nuevo" una T con un color predeterminado (rojo)
    elif i=='C': #si i es igual a C entonces
        nuevo +="\033[1;35m"+'G' # se generará en "nuevo" una G con un color predeterminado (lila)
    elif i=='T':  ##si i es igual a T entonces
        nuevo +="\033[1;34m"+'A' # se generará en "nuevo" una A con un color predeterminado (azul)
    else:
        nuevo +="\033[1;32m"+'C' # se generará en "nuevo" una C con un color predeterminado (verde)
        
print(nuevo)  #se imprime la cadena "nuevo" con los elementos (nucleotidos) reemplazados

f= open("duplicacion.txt", "w")  #se abre un nuevo documento "duplicaciòn" en formato "txt" en modo escritura
f.write(nuevo)  #Guardamos en el archivo "duplicacion" la nueva cadena de ADN 
f.close()  #cerramos archivo
```

# Transcripción
La transcripción es el primer paso de la expresión génica, el proceso por el cual la información de un gen se utiliza para generar un producto funcional, como una proteína. El objetivo de la transcripción es producir una copia de ARN de la secuencia de ADN de un gen.
El transcrito de ARN tiene la misma información que la cadena de ADN contraria a la molde (codificante) en el gen, pero contiene la base uracilo (U) en lugar de timina (T).

```python
f=open("duplicacion.txt","r") #abre el archivo generado en la celda anterior en modo lectura

text_r=f.read() # se lee el archivo
RNA=text_r  #creación de la variable
#ADN= input(n)
#Creación de tres cadenas vacías que se llenarán con los siguientes tres ciclos for
nuevo=" "
transc=""
helice=""
for i in RNA: #Para cada i en RNA
    if i=='A': #si i es igual a A entonces
        nuevo +="\033[1;33m"+'U' #en nuevo se genera una U (amarilla)
    elif i=='C': # i es igual a C entonces
        nuevo +="\033[1;35m"+'G' #en nuevo se genera una G (lila)
    elif i=='T': # i es igual a T entonces
        nuevo +="\033[1;34m"+'A' #en nuevo se genera una A (azul)
    elif i=='G': # i es igual a G entonces
        nuevo +="\033[1;32m"+'C' #en nuevo se genera una C (verde)
for i in RNA: #Para cada i en RNA
    if i=='A': #si i es igual a A entonces
        transc +='U' #en transc se reemplaza por una U
    elif i=='C': #i es igual a C entonces
        transc +='G' #en transc se reemplaza por una G
    elif i=='T': #i es igual a T entonces
        transc +='A' #en transc se reemplaza por una A
    elif i=='G': #i es igual a G entonces
        transc +='C' #en transc se reemplaza por una C
for i in RNA: #Para cada i en RNA
    if i=='A': #si i es igual a A entonces
        helice +=' A U' #en helice se reemplaza por A U
    elif i=='C': # i es igual a C entonces
        helice +=' C G' #en helice se reemplaza por C G
    elif i=='T': # i es igual a T entonces
        helice +=' T A' #en helice se reemplaza por T A
    elif i=='G': # i es igual a G entonces
        helice +=' G C' #en helice se reemplaza por G C
print(nuevo) #se imprime la cadena "nuevo"

t= open("transcripcion.txt", "w")  #abre un archivo llamado "transcripciòn" en modo escritura
t.write(transc)  #Guarda en el archivo "transcripción" la nueva cadena de ADN 
t.close() #cierra archivo
hel= open("helice.txt", "w") #abre archivo llamado "helice" en modo escritura
hel.write(helice) #guarda en el archivo helice la nueva cadena
hel.close() #cierra archivo
```

# Traducción
Durante la traducción, una célula "lee" la información contenida en el ARN mensajero (ARNm) y la usa para construir una proteína.
En un ARNm, las instrucciones para construir un polipéptido son los nucleótidos de ARN (A, U, C, y G), que se leen en grupos de tres. Estos grupos de tres se conocen como codones, el proceso se lleva a cabo con una tabla de codones que se realizará con un diccionario.
La colección de relaciones codón-aminoácidos se llama el código genético , porque permite que las células "decodifiquen" un ARNm en una cadena de aminoácidos.

```python
h=open("transcripcion.txt","r") #Abre el archivo generado en la celda de arriba en modo lectura

text=h.read() #lee el archivo
rna=text # creación de la variable
cadena=rna.replace(' ','') # cadena que quita los espacios a la cadena de RNA
def transcripcion(seq):  #definimos la funcion 
    table = { 
        'UUU':'Phe', 'UCU':'Ser', 'UAU':'Tyr', 'UGU':'Cys', 
        'UUC':'Phe', 'UCC':'Ser', 'UAC':'Tyr', 'UGC':'Cys', 
        'UUA':'Leu', 'UCA':'Ser', 'UAA':'END', 'UGA':'END',   #creamos una tabla codon (en diccionario) que se encuentra en los libros de biologia, con la que se determina a que aminoacido corresponde cada triplete de bases nitrogendas
        'UUG':'Leu', 'UCG':'Ser', 'UAG':'END', 'UGG':'Trp',                  
        'CUU':'Leu', 'CCU':'Pro', 'CAU':'His', 'CGU':'Arg', 
        'CUC':'Leu', 'CCC':'Pro', 'CAC':'His', 'CGC':'Arg', 
        'CUA':'Leu', 'CCA':'Pro', 'CAA':'Gln', 'CGA':'Arg', 
        'CUG':'Leu', 'CCG':'Pro', 'CAG':'Gln', 'CGG':'Arg', 
        'AUU':'Ile', 'ACU':'Thr', 'AAU':'Asn', 'AGU':'Ser', 
        'AUC':'Ile', 'ACC':'Thr', 'AAC':'Asn', 'AGC':'Ser', 
        'AUA':'Ile', 'ACA':'Thr', 'AAA':'Lys', 'AGA':'Arg', 
        'AUG':'Met', 'ACG':'Thr', 'AAG':'Lys', 'AGG':'Arg', 
        'GUU':'Val', 'GCU':'Ala', 'GAU':'Asp', 'GGU':'Gly', 
        'GUC':'Val', 'GCC':'Ala', 'GAC':'Asp', 'GGC':'Gly', 
        'GUA':'Val', 'GCA':'Ala', 'GAA':'Glu', 'GGA':'Gly', 
        'GUG':'Val', 'GCG':'Ala', 'GAG':'Glu', 'GGG':'Gly', 
    } 
    cadena
    proteina =""     #se crea una cadena "proteina" vacía
    if len(seq)%3 == 0: #va a ir leyendo de tres en tres la cadena de dna
        for i in range(0, len(seq), 3):  #se leerá de 3 en 3
            codon = seq[i:i + 3]    #cada triada de letras se va a buscar en el diccionario 
            proteina+= table[codon] #y se va a extraer el aminoacido correspondiente a esa triada, y los va a guardar en cadena proteina =""
    return proteina #nos devolverá nuestra cadena llena de codones
prot=transcripcion(cadena) #se define la variable
print(prot) #imprime la cadena de aminoácidos
p= open("Proteinas.txt", "w")  #abre un nuevo archivo "proteínas" en modo escritura
p.write(prot)  #Guardamos en el archivo "proteinas" la nueva cadena de aminoácidos 
p.close() #cerramos archivo
```

### Calculo de la longitud del DNA

Una molécula de ADN tiene la forma de una doble helice. El radio de cada helice es alrededor de 10 angstroms (1A=10^8). Cada helice se eleva unos 34 A cada vuelta completa y se tiene que hay aprox. 2.9x10^8 vueltas completas,con lo anterior nos dio curiosidad calcular la longitud de una helice del DNA

```python
(h1**2 + h2**2 + h3**2)**1/2
```
